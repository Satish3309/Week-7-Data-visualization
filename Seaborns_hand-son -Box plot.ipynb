#Summary 

How to Read the Boxplot

Box â†’ Shows the interquartile range (IQR: 25thâ€“75th percentile).
Line inside box â†’ Median score.
Whiskers â†’ Range of most data points (excluding outliers).
Dots outside whiskers â†’ Outliers (students with unusually high/low percentages).
Hue (gender) â†’ Each board category has two boxes side by side (Male vs Female)

Interpretation

Board-wise comparison

You can see how average percentages differ between boards (e.g., Central vs Others).
If one boardâ€™s box is consistently higher, students from that board tend to score better.

Gender comparison within boards

If the male and female boxes overlap heavily â†’ little difference in performance.
If one genderâ€™s median is consistently higher â†’ that gender tends to score better in that board.

Spread of scores

Wide boxes/long whiskers â†’ more variability in scores.
Narrow boxes â†’ scores are more consistent.

Outliers

Dots above/below whiskers show exceptional performers or very low scorers.
Useful to spot unusual cases.

```python
sb.pointplot(x = "ssc_p",y="hsc_b",data =dataset,hue='gender')
plt.show()
```

x = "ssc_p" â†’ Secondary school percentage (numeric).

y = "hsc_b" â†’ Higher secondary board (categorical).

hue = "gender" â†’ Splits each point by male vs female.

# How Pointplot Works
Each dot â†’ shows the mean value of ssc_p for that category (hsc_b).

Vertical lines (error bars) â†’ show confidence intervals (by default 95%).

Hue â†’ creates separate points for male and female within each board category.

Trend line â†’ connects the points to highlight differences across categories.

# Interpretation
Board comparison

For each hsc_b (Central, Others, etc.), you see the average SSC percentage.

If one boardâ€™s points are consistently higher, students from that board scored better in SSC.

Gender comparison

Within each board, compare male vs female points.

the female point is higher â†’ so, females scored better in SSC for that board.

If they overlap â†’ little/no difference.

Error bars

Short error bars â†’ consistent scores (low variation).

Long error bars â†’ more variability in SSC percentages.

```python
sb.boxplot(data = dataset, orient = "h")
plt.show()
```

# How to Read a Horizontal Boxplot

Box (rectangle) â†’ shows the middle 50% of values (interquartile range).

Line inside box â†’ median value.

Whiskers (horizontal lines) â†’ extend left and right, showing the spread of most data.

Dots beyond whiskers â†’ outliers (unusually high or low values).

Since orientation is "h", everything is flipped sideways:

Whiskers stretch left and right.

Median line is vertical inside the box.

# Interpretation

This plot lets you quickly compare the spread and center of all numeric features in your dataset:

For example, salary will likely show a long right whisker (because some students have very high salaries compared to the median).

Percentages (ssc_p, hsc_p, etc.) will have shorter whiskers, since scores are naturally bounded between 0â€“100.

Outliers (dots) highlight unusual cases, like extremely high or low percentages or salaries.

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('exercise')
sb.catplot(x = "time", y = "pulse", hue = "kind", kind = 'violin', col = "diet", data = df);
plt.show()
```

# What This Plot Shows

Dataset: exercise (built-in Seaborn dataset with exercise type, pulse rate, time, diet).

x = "time" â†’ Categories of exercise duration (1 min, 15 min, 30 min).

y = "pulse" â†’ Numeric variable (pulse rate of participants).

hue = "kind" â†’ Splits each violin by exercise type (rest, walking, running).

kind = "violin" â†’ Draws a violin plot (distribution shape of pulse values).

col = "diet" â†’ Creates separate subplots for each diet group (no fat, low fat).

# Interpreting the Violin Plot

Shape of violin â†’ Shows the distribution of pulse values.

Wide parts = more participants with those pulse values.

Narrow parts = fewer participants.

Median line inside violin â†’ Central tendency of pulse.

Split by hue (kind) â†’ Each violin is divided into two halves (e.g., left = male, right = female OR here = different exercise kinds).

Columns (diet) â†’ Compare distributions across diet groups side by side.
