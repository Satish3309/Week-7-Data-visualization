# Univarite Analysis

```python
sb.distplot(dataset['ssc_p'],hist=True,kde=True,kde_kws={'color':'red'})
```

# density plot details.

# hist=True
    # Tells Seaborn to display the histogram of the data.
    # Each bar shows how many students fall into that score range.
# kde=True
    # Adds a KDE curve (a smooth probability density curve) on top of the histogram.
    # This helps visualize the distribution trend more smoothly.
# kde_kws={'color':'red'}
    # Customizes the KDE curve.
    # Here, the KDE line will be drawn in red.
# x-axis = numerical column 
# y-axis = density 
# pin shape = histogram of frequency of the column "ssc_p"
# curve = density curve 
# peak means hig density

    # On the diagonal plots (where a variable is plotted against itself), instead of histograms,
        # Seaborn will show Kernel Density Estimate (KDE) curves.
    # KDE gives a smooth estimate of the distribution.

# For Categorical Plot

```python
sb.stripplot(x = "ssc_b", y = "ssc_p", data = dataset)
plt.show() 
```

# Interpretation

#Stripplot
    # A categorical scatter plot in Seaborn.
    # It plots individual data points along a categorical axis.
    # Useful for visualizing the distribution of values across categories.
    # One column is categorical and one column is numerical.
    
# Plot shows us
    # the dot represents one student's score and its grouped by school board.
    # its looks like a strip of points.
    # If many points overlap, theyâ€™ll stack vertically,
        # giving a sense of density.
    

```python
sb.stripplot(x = "ssc_b", y = "ssc_p", data = dataset,jitter=True)
plt.show()
```

# This plot shows horizontaly because jitter is True.
# jitter=True
    # Adds random noise to the horizontal position of points.
    # This prevents overlapping points from stacking directly 
        # on top of each other, making the distribution clearer.

```python
sb.stripplot(x = "gender", y = "ssc_p", data = dataset,jitter=True)
plt.show()
```

```python
sb.stripplot(x = "gender", y = "ssc_p", data = dataset)
plt.show()
```

```python
sb.swarmplot(x = "gender", y = "ssc_p", data = dataset)
plt.show()
```

# swarmplot 
    # Creates a categorical scatter plot similar to stripplot, but with one key difference:
    # Instead of adding random noise (jitter), swarmplot arranges points 
        # systematically so they donâ€™t overlap.
    # This makes the plot cleaner and more precise.

# Distribution Observation

```python
sb.violinplot(x = "gender", y = "ssc_p", data = dataset)
plt.show()
```

# violin plot, which is a combination of a boxplot and a KDE (Kernel Density Estimate) plot.

    # It shows both the summary statistics (like median, quartiles) and the distribution shape of the data.

#For each gender category:
    # The white dot â†’ median score.
    # The thick black bar â†’ interquartile range (middle 50% of scores).
    # The thin black line â†’ overall range (excluding outliers).
    # The violin shape â†’ KDE curve mirrored on both sides, showing the distribution density of scores.
    # Wider sections = more students with those scores.
    # Narrow sections = fewer students.

```python
sb.violinplot(x = "specialisation", y = "salary", data = dataset,hue="gender")
plt.show()
```
